```python
import os
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers

input_path = '../input/loan-prediction-problem-dataset/train_u6lujuX_CVtuZ9i.csv'

df = pd.read_csv(input_path)
```

```python
df = df.drop('Loan_ID',axis=1)
df.dropna(axis=0,inplace = True)
df['Income'] = df['ApplicantIncome'] + df['CoapplicantIncome']  
df.drop(['ApplicantIncome','CoapplicantIncome'],axis=1,inplace=True)
```

```python
def one_hot_encoding(df):
    for i in df.columns:
        if df[i].dtype == 'O':
            cat = set(df[i])
            cat = list(cat)
            for j in range(len(cat)):
                print(cat[j],j)
                df[i].replace(cat[j],j,inplace=True)
    
    return df
```

```python
df = one_hot_encoding(df)
```

```python
k = list(df.columns[:-2])
k.append('Income')
X= df[k].values
Y = df[df.columns[-2]].values
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.15)
```

```python
from sklearn.metrics import accuracy_score
LR = LogisticRegression()
LR.fit(X_train,Y_train)
y_hat = LR.predict(X_test)
print('Accuracy Score', accuracy_score(y_hat,Y_test))
```

```python
from sklearn.naive_bayes import GaussianNB
GB = GaussianNB()
GB.fit(X_train,Y_train)
yhat= GB.predict(X_test)
print("Accuracy Score = ",accuracy_score(yhat,Y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(X_train,Y_train)
yhat = tree.predict(X_test)
print("Accuracy Score = ",accuracy_score(yhat,Y_test))
```
